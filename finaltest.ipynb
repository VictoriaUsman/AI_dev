```python
import pandas as pd
import numpy as np
```

```python
def load_data(filepath: str) -> pd.DataFrame:
    """Load CSV file into a DataFrame and print a quick summary."""
    df = pd.read_csv(filepath)
    print(f"Loaded '{filepath}'")
    print(f"  Shape : {df.shape}")
    print(f"  Dtypes:\n{df.dtypes}\n")
    return df
```

```python
load_data('example_dataset.csv')
```

```python
def handle_missing_values(df: pd.DataFrame) -> pd.DataFrame:
    """
    Report and handle missing values.
    - Numeric columns: fill with column median.
    - If no missing values, report that and return unchanged.
    """
    missing = df.isnull().sum()
    total_missing = missing.sum()

    if total_missing == 0:
        print("Missing values: none found — no imputation needed.")
    else:
        print(f"Missing values detected:\n{missing[missing > 0]}")
        numeric_cols = df.select_dtypes(include="number").columns
        df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())
        print(f"  → Filled {total_missing} missing value(s) with column medians.")

    return df
```

```python
handle_missing_values(load_data('example_dataset.csv'))
```

```python
def remove_duplicates(df: pd.DataFrame) -> pd.DataFrame:
    """Remove exact duplicate rows and report how many were dropped."""
    before = len(df)
    df = df.drop_duplicates()
    removed = before - len(df)

    if removed == 0:
        print("Duplicates: none found — no rows removed.")
    else:
        print(f"Duplicates: removed {removed} row(s). Rows remaining: {len(df)}")

    return df.reset_index(drop=True)



```

```python
remove_duplicates(load_data('example_dataset.csv'))
```

```python
def preprocess(df: pd.DataFrame) -> pd.DataFrame:
    """Run the full cleaning pipeline."""
    print("\n--- Data Cleaning ---")
    df = handle_missing_values(df)
    df = remove_duplicates(df)
    return df
```

```python
preprocess(load_data('example_dataset.csv'))
```

```python
def add_row_statistics(df: pd.DataFrame, feature_cols: list) -> pd.DataFrame:
    """Row-wise statistics across all feature columns."""
    df["row_mean"]   = df[feature_cols].mean(axis=1)
    df["row_std"]    = df[feature_cols].std(axis=1)
    df["row_min"]    = df[feature_cols].min(axis=1)
    df["row_max"]    = df[feature_cols].max(axis=1)
    df["row_range"]  = df["row_max"] - df["row_min"]   # spread
    return df
```

```python
add_row_statistics(load_data('example_dataset.csv'), ['feature_1'])
```

```python
def add_interaction_terms(df: pd.DataFrame, feature_cols: list) -> pd.DataFrame:
    """Pairwise product interaction terms between features."""
    for i in range(len(feature_cols)):
        for j in range(i + 1, len(feature_cols)):
            a, b = feature_cols[i], feature_cols[j]
            col_name = f"{a}_x_{b}"
            df[col_name] = df[a] * df[b]
    return df
```

```python
def add_ratio_features(df: pd.DataFrame, feature_cols: list) -> pd.DataFrame:
    """
    Ratio of consecutive features (f1/f2, f2/f3, …).
    Uses a small epsilon to avoid division by zero.
    """
    eps = 1e-9
    for i in range(len(feature_cols) - 1):
        a, b = feature_cols[i], feature_cols[i + 1]
        df[f"{a}_div_{b}"] = df[a] / (df[b] + eps)
    return df
```

```python
def save_data(df: pd.DataFrame, output_path: str) -> None:
    df.to_csv(output_path, index=False)
    print(f"\nProcessed dataset saved -> '{output_path}'")
    print(f"  Shape: {df.shape}")
```

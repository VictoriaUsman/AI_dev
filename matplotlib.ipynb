```python
import matplotlib.pyplot as plt

# Sample data (10 points)
x = list(range(1, 11))          # [1, 2, ..., 10]
y = [2, 4, 5, 7, 8, 10, 9, 12, 14, 15]

# Create the plot
plt.plot(x, y, marker='o')

# Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Simple Line Graph")

# Show the graph
plt.show()
```

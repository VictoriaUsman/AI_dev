```python
def tokenize_text(paragraph):
    # Tokenize into words (split by whitespace)
    words = paragraph.split()

    # Tokenize each word into characters
    characters = [list(word) for word in words]

    return words, characters


# Example usage
text = "Hello world! This is a simple tokenizer."

words, chars = tokenize_text(text)

print("Words:")
print(words)

print("\nCharacters:")
print(chars)
```
